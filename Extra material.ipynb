# Finite difference methods

```python
dx = 0.1
x = np.arange(0, 10+dx, dx)
f = x**3

# First derivative

dfdx = np.zeros(len(f))

# First step with forward difference
dfdx[0] = (f[1] - f[0])/dx 

# Mid steps with central difference
for i in range(1, len(f)-1):
    dfdx[i] = (f[i+1] - f[i-1])/(2*dx)

# Last step with backward difference    
dfdx[-1] = (f[-1] - f[-2])/dx 

plt.plot(x, 3*x**2, label="$f'(x) = 3x^2$")
plt.plot(x[::10], dfdx[::10], marker='o', ls='', label='Numerical')
plt.xlabel('x')
plt.legend()
plt.show()
```

# Root finding

### Relaxation Method for non-linear equations

We aim to solve

\begin{equation}
x + e^{-2x} = 1.
\end{equation}

We could rewrite this equation in the form of $x = f(x)$ and iterate until we find a convergent $x_0$ value. One possibility is

\begin{equation}
x  = 1 + e^{-2x}, 
\end{equation}
in this form, $f(x) = 1 + e^{-2x}$.

```python
# Inspecting the problem graphically

def f(x):
    return 1 - np.exp(-2*x)

x = np.linspace(-1, 1, 101)

plt.plot(x, f(x), label='f(x)')
plt.plot(x, x, label='x')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```

```python
# Writting our relaxation function

def relaxation_method(f, x0, N=100, tol=1e-6):
    x_old = x0
    for i in range(N):
        x_new = f(x_old)
        if abs(x_new - x_old) < tol:
            print(f'Converged at x = {x_new} after {i+1} iterations.')
            break
        x_old = x_new
    else:
        print(f'Convergence failed. Last x = {x_new}.')
```

```python
# Looks like we have a solution around x = 0.5

relaxation_method(f, 0.5)
```

```python
# The other solution seems to be  around x = 0. But that won't converge with our choice for f(x)

relaxation_method(f, -0.1)
```

Instead, we could make a different choice for f(x):

\begin{equation}
x + e^{-2x} = 1.
\end{equation}

\begin{equation}
x = -\frac{1}{2} \ln(1 - x), 
\end{equation}

in this form, $f(x) = -\frac{1}{2} \ln(1 - x)$.

```python
# Inspecting the problem graphically

def f_new(x):
    return -0.5*np.log(1-x)

x = np.linspace(-0.9, 0.9, 101)

plt.plot(x, f_new(x), label='f(x)')
plt.plot(x, x, label='x')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.show()
```

```python
# Solution around 0.7

relaxation_method(f_new, 0.01)
```

Why sometimes our method doesn't converge? Let our solution be $x_0 = f(x_0)$ at the $i^{th}$ iteration

\begin{equation}
x_i = f(x_{i-1}).
\end{equation}

Taylor expanding the RHS around $x_0$:

\begin{equation}
x_i = f(x_0) + (x_{i-1} - x_0)f'(x_0) + ...
\end{equation}

but $f(x_0) = x_0$. Defining $\Delta x_i = (x_i - x_0)$,

\begin{equation}
x_i - x_0 \approx (x_{i-1} - x_0)f'(x_0)
\end{equation}

\begin{equation}
\Delta x_i \approx \Delta x_{i-1}f'(x_0)
\end{equation}

\begin{equation}
\frac{\Delta x_i}{\Delta x_{i-1}} \approx f'(x_0)
\end{equation}

Therefore, our result will converge if $|f'(x_0)| < 1$.

# Many particle simulation

```python
dt = 0.01
t_span = np.arange(0, 15 + dt, dt)
L = 15.0

N_times = len(t_span)
N_dimensions = 2
N_particles = 20

m = np.ones(N_particles)

r = np.zeros((N_times, N_particles, N_dimensions)) 
v = np.zeros((N_times, N_particles, N_dimensions))
a = np.zeros((N_times, N_particles, N_dimensions))

# Staring positions and velocities

r[0] = 7*np.random.randn(N_particles, N_dimensions) + 0.0

for i in range(N_particles):
    for j in range(N_particles):
        if i != j:
            rij = np.linalg.norm(r[0][i] - r[0][j])
            if rij < 0.5:
                r[0][j] = r[0][j] + np.random.random(2)

v[0] = np.random.randn(N_particles, N_dimensions)

fig = plt.figure(figsize=(6,6))
plt.plot(r[0][:,0], r[0][:,1], marker='o', ls='')
plt.xlim(-L,L)
plt.ylim(-L,L)
plt.show()
```

```python
# Running

def calculate_lj_acceleration(t, r, eps=1e-3, sigma=0.5):
    for i in range(N_particles):
        for j in range(N_particles):
            if i !=j:
                rij = r[t, i] - r[t, j]
                rij_norm = np.linalg.norm(rij)
                fij = (48*eps/rij_norm**2)*((sigma/rij_norm)**12 - 0.5*(sigma/rij_norm)**6)*rij/m[i] 
                a[t, i] = a[t, i] + fij

                               
for t in range(N_times-1):
    calculate_lj_acceleration(t, r)
    v[t+1] = v[t] + a[t]*dt
    r[t+1] = r[t] + v[t+1]*dt
    # PBC
    r[t+1] = np.where(r[t+1] > L, r[t+1] - 2*L, r[t+1])
    r[t+1] = np.where(r[t+1] < -L, r[t+1] + 2*L, r[t+1])
```

```python
# Animating

fig = plt.figure(figsize=(6,6))
lines = plt.plot([], marker='o', ls='')
line = lines[0]
plt.xlim(-L, L)
plt.ylim(-L, L)

def animate(frame):
    line.set_data((r[frame][:,0], r[frame][:,1]))
    
anim = FuncAnimation(fig, animate, frames=N_times, interval=dt*1000)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
```
